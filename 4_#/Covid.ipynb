```python
#Q. 1) Show the number of Confirmed, Deaths and Recovered cases in each Region.


Data.groupby(['Region'])['Confirmed','Deaths','Recovered'].sum()
```

```python
#Q. 2) Remove all the records where the Confirmed Cases is Less Than 10.
print('Before:',Data.shape)
less_than_10=Data[Data['Confirmed']<10].index
Data.drop(less_than_10,inplace=True,axis=0)
print('After:',Data.shape)
```

```python
#Q. 3) In which Region, maximum number of Confirmed cases were recorded ?

Max_Confirm=Data.groupby(['Region'])['Confirmed'].sum().sort_values(ascending=False).index[0]
print('maximum number of Confirmed cases recorded in:=',Max_Confirm)
```

```python
#Q. 4) In which Region, minimum number of Deaths cases were recorded ?
```

```python
Temp_Data=Data.groupby(['Region'])['Deaths'].sum()
Temp_Data[Temp_Data==1].sort_values(ascending=True)
```
